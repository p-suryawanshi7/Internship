## Q.10.Answer: A python program to scrape the salary data for Data Scientist designation with Company name, Number of salaries, Average salary, Minsalary, Max Salary

```python
driver10=webdriver.Chrome('chromedriver.exe')
driver10.get('https://www.ambitionbox.com/')
```

```python
driver10.maximize_window()
```

```python
salary=driver10.find_element(By.XPATH,'/html/body/div[1]/nav[2]/div/ul/li[3]/a')
salary.click()
```

```python
browser=driver10.find_element(By.XPATH,'/html/body/div[1]/nav[2]/div/ul/li[3]/div/ul/li[1]/div/div[2]/p')
browser.click()
```

```python
job=driver10.find_element(By.XPATH,'//input[@class="tt-input"]')
job.send_keys('Data Scientist')
```

```python
select=driver10.find_element(By.XPATH,'/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div')
select.click()
```

```python
comp=[]
for i in driver10.find_elements(By.XPATH,'//div[@class="company-info"]/a'):
    comp.append(i.text.split("\n"))
    
company=sum(comp,[])
del company[1:20:2]

bsal=[]
for a in driver10.find_elements(By.XPATH,'//div[@class="sbold-list-header"]/span'):
    bsal.append(a.text.replace('(','').replace(')',''))

avgsal=[]
for b in driver10.find_elements(By.XPATH,'//p[@class="averageCtc"]'): 
    avgsal.append(b.text)

min_sal=[]
max_sal=[]
for c in driver10.find_elements(By.XPATH,'//div[@class="value body-medium"]'): 
    min_sal.append(c.text)
    max_sal.append(c.text)
del min_sal[1:20:2]
del max_sal[0:20:2]

exp=[]
for d in driver10.find_elements(By.XPATH,'//div[@class="sbold-list-header"]'): 
    exp.append(d.text.split('('))
    
req_exp=sum(exp,[]) 
del req_exp[1:20:2]

df=pd.DataFrame({'Company Name':company,'Total Salary Record':bsal,'Average Salary':avgsal,'Minimum Salary':min_sal,'Maximum Salary':max_sal,'Experience Required':req_exp})
df
```
