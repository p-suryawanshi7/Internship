```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
import time

import warnings
warnings.filterwarnings('ignore')
```

## 1. Scraping data from "Cardekho" Website

### Ahmedabad

```python
driver1=webdriver.Chrome('chromedriver.exe')
url1=("https://www.cardekho.com/used-cars")
driver1.get(url1)
driver1.maximize_window()
time.sleep(2)
```

```python
location_Ahmedabad= driver1.find_element(By.XPATH,"//a[@title='Used cars in Ahmedabad']").click()
```

```python
# scrolling the web page 
for _ in range(300):
    driver1.execute_script("window.scrollBy(0,100)")
```

```python
urls = []
for i in range(20):
    for j in driver1.find_elements(By.XPATH,'//div[@class="gsc_col-xs-7  carsName"]/a'):
        urls.append(j.get_attribute('href'))

time.sleep(2)
print(len(urls))
```

```python
#Creating empty list 
Brand= []
Model = []
Variant=[]
Manufacturing_Year = []
Driven_KiloMeters= []
Fuel = []
Location= []
Car_Price= []

#Fetching data for Brand    
try:
    Brand_name=driver1.find_elements(By.XPATH,"//div[@class='gsc_col-xs-7  carsName']/a")
    for i in Brand_name:
        Brand.append(i.text.split(" ")[1])
except NoSuchElementException as e:
    Brand.append('-')

#Fetching data for Model   
try:
    Model_name=driver1.find_elements(By.XPATH,"//div[@class='gsc_col-xs-7  carsName']/a")
    for i in Model_name:
        Model.append(i.text[5:])
except NoSuchElementException as e:
    Model.append('-')

#Fetching data for Variant   
try:
    Variant_name=driver1.find_elements(By.XPATH,"//div[@class='car_variant']")
    for i in Variant_name:
        Variant.append(i.text)
except NoSuchElementException as e:
    Variant.append('-')

#Fetching data for Manufacturing    
try:
    Manufacturing=driver1.find_elements(By.XPATH,"//div[@class='gsc_col-xs-7  carsName']/a")
    for i in Manufacturing:
        Manufacturing_Year.append(i.text.split(" ")[0])
except NoSuchElementException as e:
    Manufacturing_Year.append('-') 

#Fetching data for Driven_KiloMeters    
try:
    Driven_KiloMeter=driver1.find_elements(By.XPATH,'//div[@class="truncate dotlist-2"]/div[1]')
    for i in Driven_KiloMeter:
        Driven_KiloMeters.append(i.text)
except NoSuchElementException as e:
    Driven_KiloMeters.append('-')    

#Fetching data for Fuel    
try:
    Fuel_name=driver1.find_elements(By.XPATH,'//div[@class="truncate dotlist-2"]/div[2]')
    for i in Fuel_name:
        Fuel.append(i.text)
except NoSuchElementException as e:
    Fuel.append('-')

#Fetching data for Location   
try:
    Location_name=driver1.find_elements(By.XPATH,'//div[@class="gsc_col-xs-12 mainheading"]')
    for j in range (200):
        for i in Location_name:
            Location.append(i.text[13:])
except NoSuchElementException as e:
    Location.append('-')

#Fetching data for Car_Price    
try:
    Car_Pr=driver1.find_elements(By.XPATH,"//span[@class='amnt ']")
    for i in Car_Pr:
        Car_Price.append(i.text)
except NoSuchElementException as e:
    Car_Price.append('-')    
```

```python
print(len(Brand),len(Model),len(Variant),len(Manufacturing_Year), len(Driven_KiloMeters), len(Fuel),  len(Location), len(Car_Price))        
```

```python
Ahmedabad=pd.DataFrame({'Brand':Brand[:160], 'Model':Model[:160], 'Variant':Variant[:160], 'Manufacturing_Year':Manufacturing_Year[:160], 'Driven_KiloMeters':Driven_KiloMeters[:160], 'Fuel':Fuel[:160],  'Location':Location[:160], 'Car_Price':Car_Price[:160]})
Ahmedabad
```

```python
driver1.close()
```
